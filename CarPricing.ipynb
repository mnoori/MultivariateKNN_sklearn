```python
import pandas as pd
import numpy as np

cars=pd.read_csv('imports-85.data')
cars.columns
```

This does not match with what that has been mentioned in teh dataset documentation. We will take the column names from the documentation.

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars=pd.read_csv('imports-85.data',names=cols)
cars.info()
```

We select the columns with numerical values.

```python
numerical_cols=['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars=cars[numerical_cols]
print('shape of data set is:',numeric_cars.shape)
numeric_cars.head()
```

### Data Cleaning 

let's replace the `?` with `Nan` 

```python
numeric_cars = numeric_cars.replace('?', np.nan)
numeric_cars
```

```python
numeric_cars=numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

First, since `price` is our target variable, we will remove any rows that its `price` is missing.
For the rest of columns, we will replace the missing values with columns means.

```python
numeric_cars=numeric_cars.dropna(subset=['price'])
numeric_cars=numeric_cars.fillna(numeric_cars.mean())

#confirming there is no missing values.
numeric_cars.isnull().sum()
```

##### Normalizing columns (except price)

```python
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
```

## Univariate Model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(train_col, taregt_col, df):
    model=KNeighborsRegressor()
    np.random.seed(1)
    
    #shuffling the deck!
    shuffled_index=np.random.permutation(df.index)
    rand_df=df.reindex(shuffled_index)
    
    #last train row
    train_endrow=int(len(rand_df) / 2)
    
    #creating train and test datasets
    train_df=rand_df.iloc[:train_endrow]
    test_df=rand_df.iloc[train_endrow:]
    
    #model fitting
    model.fit(train_df[[train_col]],train_df[taregt_col])
    predictions=model.predict(test_df[[train_col]])
    
    #rmse
    mse=mean_squared_error(test_df[taregt_col],predictions)
    rmse=np.sqrt(mse)
    
    return rmse

rmse_results={}
train_cols=numeric_cars.columns.drop('price') #take all columns except `price`

#for all columns, fit the model seperately and report the rmse
for col in train_cols:
    rmse_results[col]=knn_train_test(col,'price',numeric_cars)

# create a series object from dictionary, so we can sort and see the results easily

rmse_series=pd.Series(rmse_results)
rmse_series.sort_values()
```

Now, let' modify k in K nearest neighbors method. We will test the model using different k values.

```python
def knn_train_test(train_col, taregt_col, df):
    np.random.seed(1)
    
    #shuffling the deck!
    shuffled_index=np.random.permutation(df.index)
    rand_df=df.reindex(shuffled_index)
    
    #last train row
    train_endrow=int(len(rand_df) / 2)
    
    #creating train and test datasets
    train_df=rand_df.iloc[:train_endrow]
    test_df=rand_df.iloc[train_endrow:]
    
    k_values=[(i*2+1) for i in range(5)]
    k_rmse={}
    for k in k_values:
        model=KNeighborsRegressor(n_neighbors=k)
        #model fitting
        model.fit(train_df[[train_col]],train_df[taregt_col])
        predictions=model.predict(test_df[[train_col]])
        
        #rmse
        mse=mean_squared_error(test_df[taregt_col],predictions)
        k_rmse[k]=np.sqrt(mse)
    
    return k_rmse

k_rmse_results={}
train_cols=numeric_cars.columns.drop('price') #take all columns except `price`

#for all columns, fit the model seperately and report the rmse
for col in train_cols:
    k_rmse_results[col]=knn_train_test(col,'price',numeric_cars)

# create a series object from dictionary, so we can sort and see the results easily

k_rmse_pd=pd.DataFrame(k_rmse_results)
k_rmse_pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig,ax=plt.subplots(figsize=(12,8))

k_rmse_pd.plot(ax=ax)
plt.xlabel('k value')
plt.ylabel('RMSE')
plt.legend(loc='upper right')
```

```python
def knn_train_test(train_matrix, taregt_col, df):
    np.random.seed(1)
    
    #shuffling the deck!
    shuffled_index=np.random.permutation(df.index)
    rand_df=df.reindex(shuffled_index)
    
    #last train row
    train_endrow=int(len(rand_df) / 2)
    
    #creating train and test datasets
    train_df=rand_df.iloc[:train_endrow]
    test_df=rand_df.iloc[train_endrow:]
    
    k_values=[i for i in range(1,25)]
    k_rmse={}
    
    for k in k_values:
        model=KNeighborsRegressor(n_neighbors=k)
        #model fitting
        model.fit(train_df[train_matrix],train_df[taregt_col])
        predictions=model.predict(test_df[train_matrix])
        
        #rmse
        mse=mean_squared_error(test_df[taregt_col],predictions)
        k_rmse[k]=np.sqrt(mse)
    
    return k_rmse

five_best_features=['highway-mpg','horsepower','city-mpg','curb-weight','length']

k_rmse_results= knn_train_test(five_best_features,'price',numeric_cars)

k_rmse_results=pd.Series(k_rmse_results)
k_rmse_results
```
